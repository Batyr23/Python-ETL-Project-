# Sales Data Exploration

**Aim**:  To analyze and derive actionable insights from sales and inventory data to optimize business performance, enhance customer experience, and improve decision-making processes.

## Tables we have:
    1.food_inventory
    2.sales_by_store
    3.store_lookup
    4.customer_lookup
    5.employee_lookup
    6.product_lookup
    



### 1. food_inventory
This table contains details about daily pastry stock level, quantity used, and waste. Columns in this table:

    1.store_id                        
    2.baked_date               
    3.transaction_date         
    4.product_id                      
    5.quantity_start_of_day           
    6.quantity_sold                   




### 2. sales_by_store
This table contains transaction data including date, store, employee, quantity sold, price, etc. Columns in this table:

    1.transaction_id            
    2.transaction_date   
    3.transaction_time   
    4.store_id                  
    5.staff_id                  
    6.customer_id               
    7.instore_yn                 
    8.order                     
    9.line_item_id              
    10.product_id                
    11.quantity_sold             
    12.unit_price                
    13.promo_item_yn



### 3. store_lookup
Details for each store location, address, phone number and store type. Columns in this table:

    1.store_id                  
    2.store_type               
    3.store_square_feet         
    4.store_address            
    5.store_city               
    6.store_state_province     
    7.store_postal_code         
    8.store_longitude         
    9.store_latitude          
    10.manager                  
    11.Neighorhood


### 4. customer_lookup
Customer ID, home store, name, email, loyalty details, birthdate, gender. Columns in this table:

    1.customer_id                     
    2.home_store                      
    3.customer_first-name            
    4.customer_email                 
    5.customer_since         
    6.loyalty_card_number            
    7.birthdate              
    8.gender                         
    9.birth_year 
    

### 5. employee_lookup
Details for each staff member, their name, position, location and start date. Columns in this table:

    1.staff_id               
    2.first_name            
    3.last_name             
    4.position              
    5.start_date    
    6.location

### 6. product_lookup
Contains details for each product like type, group, category, unit of measure, wholesale price, retail price, tax exempt, discount and new product. Columns in this table:

    1.product_id                   
    2.product_group               
    3.product_category            
    4.product_type                
    5.product                     
    6.product_description         
    7.unit_of_measure             
    8.current_cost               
    9.current_wholesale_price    
    10.current_retail_price       
    11.tax_exempt_yn               
    12.promo_yn                    
    13.new_product_yn
