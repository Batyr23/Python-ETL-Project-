# **Cleaning**

### **Basic statistic about null values and data types**

Создаем функцию, которая проверяет типы данных и наличие пустых значений в нашей таблице.

```python
def stats_null(df):
    print('Total rows: ', '\n', len(df), '\n')
    for columns in df.columns:
        null_counts = df[columns].isnull().sum()
        print(f'Null values in column "{columns}": {null_counts}')

    print('\n')
    print(df.dtypes)
```

### Tables:
##### 1. food_inventory (Эта таблица содержит информацию о ежедневном уровне запасов выпечки, использованном количестве и отходах.)

```python
stats_null(food_inventory_copy)
```

**Здесь много значений null, поэтому нам нужно их удалить или обработать.**

```python
#dropping rows with null Ids
food_inventory_copy = food_inventory_copy.dropna(subset=['store_id', 'product_id'])


#replacing date columns null by their mode value
to_date_columns = ['baked_date', 'transaction_date']
mode_value = food_inventory_copy[to_date_columns].mode().iloc[0]
food_inventory_copy.loc[:, to_date_columns] = food_inventory_copy.loc[:, to_date_columns].fillna(value=mode_value)


#changing data type
food_inventory_copy[to_date_columns] = food_inventory_copy[to_date_columns].apply(pd.to_datetime)


#replacing other columns null by their mean value
columns_to_fill = ['quantity_start_of_day', 'quantity_sold']
mean_value = food_inventory_copy[columns_to_fill].mean()
food_inventory_copy.loc[:, columns_to_fill] = food_inventory_copy.loc[:, columns_to_fill].fillna(value=mean_value)
```

##### **2. sales_by_store (Эта таблица содержит данные о транзакциях, включая дату, магазин, сотрудника, количество проданных товаров, цену и т. д.)**

```python
stats_null(sales_by_store_copy)
```

**Здесь много значений null, поэтому нам нужно их удалить или обработать.**

```python
#dropping IDs with null
sales_by_store_copy = sales_by_store_copy.dropna(subset=['transaction_id', 'store_id', 'staff_id', 'customer_id',
                                                         'line_item_id', 'product_id'])


#replacing date columns null by their mode value
to_date_columns = ['transaction_date', 'transaction_time']
mode_value = sales_by_store_copy[to_date_columns].mode().iloc[0]
sales_by_store_copy.loc[:, to_date_columns] = sales_by_store_copy.loc[:, to_date_columns].fillna(value=mode_value)


# changing data type
sales_by_store_copy[to_date_columns] = sales_by_store_copy[to_date_columns].apply(pd.to_datetime)


#replacing categorical columns null by their mode value
categorical_columns = ['promo_item_yn', 'instore_yn', 'order', 'quantity_sold']
mode_value = sales_by_store_copy[categorical_columns].mode().iloc[0]
sales_by_store_copy.loc[:, categorical_columns] = sales_by_store_copy.loc[:, categorical_columns].fillna(value=mode_value)


#replacing other nulls by their mean value
columns_to_fill = ['unit_price']
mean_value = sales_by_store_copy[columns_to_fill].mean()
sales_by_store_copy.loc[:, columns_to_fill] = sales_by_store_copy.loc[:, columns_to_fill].fillna(value=mean_value)
```

### **5) Dynamics of Food Inventory quantity sold by WeekDays: to find which day highest sold items we have**

```python
# Creating new column with Week Day
week_day = food_inventory_copy.copy(deep=True)
week_day['week'] = week_day['transaction_date'].dt.dayofweek
week_day['week_day'] = week_day['week'].map({
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday',
})

# Sorting the dataframe by 'week' to ensure correct order
week_day = week_day.sort_values('week')

# Convert 'week_day' to categorical type with desired sorting order
week_day['week_day'] = pd.Categorical(week_day['week_day'], categories=[
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    ordered=True)

# Grouping by 'week_day' and summing 'quantity_sold'
weekday_sale = week_day.groupby('week_day')['quantity_sold'].sum().reset_index()

# Plotting the bar chart
plt.figure(figsize=(7, 4))
sns.barplot(x='week_day', y='quantity_sold', data=weekday_sale, palette='viridis')
plt.title('Quantity sold by WeekDays')
plt.grid(True)
plt.show()
```

### **6) Number of products by product group from Product lookup table with Actually sold products from sales.**

```python
product_count = product_sales.groupby('product_group')['quantity_sold'].sum().reset_index()

# Sort the data in descending order
product_count_sorted = product_count.sort_values(by='quantity_sold', ascending=False)

# Plot the pie chart
plt.figure(figsize=(6, 6))
patches, labels, autopcts = plt.pie(product_count_sorted['quantity_sold'], startangle=90, autopct='%1.1f%%')

# Extract the autopct values from the pie chart
autopct_values = [float(autopct.get_text().strip('%')) for autopct in autopcts]

# Construct legend labels with both product group names and their corresponding percentage values
legend_labels = [f"{group}: {quantity:.1f}%" for group, quantity in zip(product_count_sorted['product_group'], autopct_values)]

# Add legend with values
plt.legend(legend_labels, loc="center left", bbox_to_anchor=(1, 0.5))

# Place labels near the pie chart
plt.title('Number of sold products by product group')
plt.tight_layout()
plt.show()
```

# **MYSQL Connection Loading Data to DB**

To connect to a MySQL database through Python and send Pandas DataFrames into the database, you can use the pandas library along with sqlalchemy, which provides a way to interact with databases using Python. You'll also need the mysql-connector-python package to connect specifically to MySQL databases. Here's a step-by-step guide:

#### **Approach to insert Dataframes as Tables with function inside MySQL DB**

```python
from sqlalchemy import create_engine, inspect, text
import pandas as pd

# Function to insert dataframe into MySQL database
import time
import os
import pandas as pd

# Function to insert dataframe into MySQL and collect metadata
def insert_dataframe_and_collect_metadata(df, table_name, engine):
    metadata = {}  # Dictionary to store metadata
    
    # Record start time
    start_time = time.time()
    metadata['start_time'] = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(start_time))
    
    # Insert data into SQL table
    try:
        df.to_sql(table_name, con=engine, if_exists="append", index=False)
        print(f"Data inserted into '{table_name}' table successfully!")
        
        # Record end time and other metadata
        end_time = time.time()
        metadata['end_time'] = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(end_time))
        metadata['duration_seconds'] = end_time - start_time
        metadata['table_name'] = table_name
        metadata['number_of_rows'] = df.shape[0]
        metadata['byte_size'] = df.memory_usage(index=True, deep=True).sum()  # Total memory in bytes
        
        return metadata
    except Exception as e:
        print(f"Failed to insert data into '{table_name}' table:", e)
        return None
```

```python
# Configure Database Connection
username = "root"
password = "root"
hostname = "localhost"  # or your host address
database_name = "pythonProject"
connection_string = f"mysql+mysqlconnector://{username}:{password}@{hostname}/{database_name}"
engine = create_engine(connection_string)

# Check Connection Status
try:
    engine.connect()
    print("Connection successful!")
except Exception as e:
    print("Connection failed:", e)
```

#### Начинаем процесс загрузки данных в MySQL DB (Нужно еще учитывать что данные загружаются по DML Insert который является append в питоне, если несколько раз одну таблицу загружать, то соответсвенно вы будете дублировать строки, для этого есть процесс так называемый deduplication в бд)

```python
# DataFrames and their corresponding SQL table names
dataframes = {
    "food_inventory_copy": food_inventory_copy,
    "sales_by_store_copy": sales_by_store_copy,
    "store_lookup_copy": store_lookup_copy,
    "customer_lookup_copy": customer_lookup_copy,
    "employee_lookup_copy": employee_lookup_copy,
    "product_lookup_copy": product_lookup_copy
}

# Collect metadata for all dataframes and write to a file
metadata_list = []

for table_name, df in dataframes.items():
    metadata = insert_dataframe_and_collect_metadata(df, table_name, engine)
    if metadata:
        metadata_list.append(metadata)

# Convert list of metadata to a DataFrame
metadata_df = pd.DataFrame(metadata_list)

# Save the metadata DataFrame to a CSV file
metadata_df.to_csv('dataframe_metadata.csv')
print(f"Metadata saved to {metadata_filename}")
```
